```python
from transformers import pipeline

# Use Flan-T5 (text2text) — stronger than GPT-2
ai_worker = pipeline("text2text-generation", model="google/flan-t5-base")

def run_task(task: str) -> str:
    result = ai_worker(task, max_length=200, do_sample=True)
    return result[0]['generated_text']

# Test it
print(run_task("Give me 3 futuristic trillion-dollar startup ideas"))
```

```python
import torch
from transformers import pipeline

# choose model (if GPU available, Colab may use it)
AGENT_MODEL = "google/flan-t5-base"   # change to "sshleifer/tiny-gpt2" if too slow or OOM

device = 0 if torch.cuda.is_available() else -1
print("Using device =", device, "  (0=GPU, -1=CPU)")

print("Loading agents (this may take ~a minute)...")
agent1 = pipeline("text2text-generation", model=AGENT_MODEL, device=device)
agent2 = pipeline("text2text-generation", model=AGENT_MODEL, device=device)
print("Agents loaded.")
```

```python
def run_agent(agent, prompt, max_new_tokens=150):
    out = agent(prompt, max_new_tokens=max_new_tokens)
    return out[0]["generated_text"].strip()

def ai_task_flow(task):
    print("=== Incoming Task ===")
    print(task, "\n")
    plan = run_agent(agent1, f"Break this task into clear numbered steps:\n\n{task}", max_new_tokens=150)
    print("Agent1 (planner) output:\n", plan, "\n")
    execution = run_agent(agent2, f"Take the plan below and produce a client-ready deliverable:\n\n{plan}", max_new_tokens=250)
    print("Agent2 (executor) output:\n", execution, "\n")
    return plan, execution

# quick smoke test:
plan, execution = ai_task_flow("Create an outline and a 300-word intro for an online beginner Python course for adults.")
```

```python
# In-memory ledger (simulate balances)
ledger = {
    "platform": 0.0,   # your cut
    "agent1": 0.0,     # earnings to agent1
    "agent2": 0.0,
    "client": 100.0    # simulate a client's balance
}

def pay_for_task(client_id, worker_id, task_price, platform_fee=0.10):
    platform_cut = round(task_price * platform_fee, 2)
    worker_receive = round(task_price - platform_cut, 2)
    ledger[client_id] = ledger.get(client_id, 0.0) - task_price
    ledger[worker_id] = ledger.get(worker_id, 0.0) + worker_receive
    ledger["platform"] = ledger.get("platform", 0.0) + platform_cut
    return platform_cut, worker_receive

# Job queue example
job_queue = [
    {"id": 1, "client_id": "client", "task": "Write a 300-word blog post on benefits of AI tutors", "price": 10.0},
    {"id": 2, "client_id": "client", "task": "Create a 7-lesson outline for Python beginners", "price": 15.0}
]

def process_job(job):
    print("Processing job:", job["id"])
    plan, execution = ai_task_flow(job["task"])
    platform_cut, worker_receive = pay_for_task(job["client_id"], "agent2", job["price"], platform_fee=0.10)
    filename = f"deliverable_job_{job['id']}.txt"
    with open(filename, "w", encoding="utf-8") as f:
        f.write("TASK:\n" + job["task"] + "\n\nPLAN:\n" + plan + "\n\nDELIVERABLE:\n" + execution)
    print(f"Job {job['id']} done. Saved to {filename}. Payment: platform ${platform_cut}, worker ${worker_receive}\n")
    return filename

# Process all jobs in the demo queue
saved_files = []
for j in job_queue:
    saved_files.append(process_job(j))

print("Final ledger:", ledger)
```

```python
import json
metrics = {
    "jobs_completed": len(job_queue),
    "platform_revenue": ledger["platform"],
    "client_balance": ledger["client"]
}
with open("ledger.json", "w", encoding="utf-8") as f:
    json.dump(ledger, f, indent=2)
with open("metrics.json", "w", encoding="utf-8") as f:
    json.dump(metrics, f, indent=2)
print("Saved ledger.json and metrics.json. Files:", saved_files)
```
